#### Run this cell to connect to your GIS and get started:

```python
from arcgis.gis import GIS
import pandas as pd
gis = GIS("home")
```

#### Now you are ready to start!

```python
def type(item):
    return f'type:"{item}"'

querystring = f'({type("Web Map")} OR {type("Web Scene")} OR {type("Dashboard")} OR {type("Form")} OR {type("Web Experience")} OR {type("Web Mapping Application")} OR {type("QuickCapture Project")} OR {type("StoryMap")}  )'
querystring1 = f'({type("StoryMap")} )'

content_MapsApps = gis.content.search(query = querystring, max_items=1000)
print(len(content_MapsApps), "...Items Queried")


df = pd.DataFrame(content_MapsApps)
df.insert(2, "Name:", df['title'] + " || " + df['type'] , True)
df_subset = df[['Name:', 'numViews']].rename(columns={'numViews': 'Views:'})
df_sorted = df_subset.sort_values('Views:', ascending=False).reset_index(drop=True)
df_sorted

df_sorted.to_excel('allMapsApps_sortedbyViews.xlsx')
print("Saved Excel File...")
```
