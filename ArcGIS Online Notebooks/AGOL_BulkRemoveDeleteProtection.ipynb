#### Run this cell to connect to your GIS and get started:

```python
from arcgis.gis import GIS
gis = GIS("home")
```

#### Now you are ready to start!

```python
serviceDefinitionArray = gis.content.search(query='', item_type='Service Definition', max_items=1000)
print(len(serviceDefinitionArray), "Service Definition Items Queried...")

for sd in serviceDefinitionArray:
    sd.protect(enable=False)

print("All service definitions delete protection have been removed.")
```
